```python
class Dog():
    def __init__(self, myBreed="Rottweiler", dogName="Micky"):
        self.breed = myBreed
        self.name = dogName
        
    def print_dog_breed(self):
        print(self.breed)
        
    def print_dog_name(self):
        print(self.name)
        
    def define_dog(self):
        print(f"My dog {self.name} breed is {self.breed}")
```

```python
rottweiler = Dog()     
rottweiler.define_dog()
```

```python
class Animal():
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound
        
    def make_sound(self):
        raise NotImplementedError("Must be created")
        
    def __str__(self):
        return f"{self.name}, {self.sound}"
```

```python
class Dog(Animal):
    
    def make_sound():
        return self.name + "woof! woof!! woof!!!"
```

```python
class Cat(Animal):
    
    def make_sound():
        return self.name + "meow! meow!! meow!!!"
```

```python
# Decorators
```

```python
def add_decorator(myFunc):
    
    def addNum(a, b):
        return myFunc(a, b)
    return addNum
```

```python
@add_decorator
def addFunc(a, b):
    return a + b
```

```python
addFunc(10,5)
```

```python
def generate_fibonacci(n):
    a = 1
    b = 1
    
    for x in range(n):
        yield a 
        a, b = b, a+b
```

```python
for num in generate_fibonacci(15):
    print(num)
```

```python
def generate_num(n):
   
    for x in range(n):
        yield x
```

```python
# Timing 
import time
import timeit
```

```python
def func_one(n):
    return [str(num) for num in range(10)]
```

```python
def func_two(n):
    return list(map(str,range(n)))
```

```python
time_start = time.time()
result = func_one(2000000)
time_end = time.time()
time_elapsed = time_end - time_start
print(time_elapsed)
```

```python
time_start = time.time()
result = func_two(2000000)
time_end = time.time()
time_elapsed = time_end - time_start
print(time_elapsed)
```

```python
stmt = """
func_one(100)
"""
setup = """
def func_one(n):
    return [str(num) for num in range(10)]
    """
timeit.timeit(stmt, setup, number=100000)
```

```python
stmt2 = """
func_two(100)
"""
setup2 = """
def func_two(n):
    return list(map(str,range(n)))
    """
timeit.timeit(stmt2, setup2, number=100000)
```

```python
%%timeit
func_one(100)
```
